```python
import pytesseract
from PIL import Image
from pdf2image import convert_from_path
import re

# Path to your scanned PDF file
pdf_file = "Pdf_1.pdf"

# Convert the PDF to a list of PIL (Pillow) images
images = convert_from_path(pdf_file)

# Initialize Tesseract with the location of the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Update with your Tesseract path

# Extract text from each page of the PDF
text = ""
for page_num, image in enumerate(images):
    text += pytesseract.image_to_string(image, lang='eng')  # Use 'eng' for English, or specify the appropriate language code

# Print or save the extracted text
print(text) 
```

```python
# Your extracted text
extracted_text = text

# Initialize a list to store the matched lines
matched_lines = []

# Use a regular expression to find and extract the relevant information
pattern = r'\d+ [A-Z0-9]+ .*? \d+\.\d{2} \d+\.\d{2}'
matches = re.findall(pattern, extracted_text)

# Add the matched lines to the list
matched_lines.extend(matches)

# Print the extracted lines
extracted_output=[]
for i, line in enumerate(matched_lines, start=1):
    extracted_output.append(f"{i}. {line}")
```

```python
import re
import pandas as pd

def process_item(item):
    # Define regular expressions to extract values
    item_line_pattern = r'^(\d+)\.'
    item_number_pattern = r'(\d+)\s([A-Z0-9]+)'
    description_pattern = r' ([^0-9]+) (\d+cm)?'
    uom_pattern = r' (EA|\S+)'
    quantity_pattern=r'(\d+) (\d+\.\d{2})'
    extended_price_pattern = r' (\d+\.\d{2})$'  

    # Extract values using regular expressions
    item_line_match = re.search(item_line_pattern, item)
    item_number_match = re.search(item_number_pattern, item)
    description_match = re.search(description_pattern, item)
    uom_match = re.search(uom_pattern, item)
    quantity_match=re.search(quantity_pattern, item)
    extended_price_match = re.search(extended_price_pattern, item)

    item_line = item_line_match.group(1) if item_line_match else None
    item_number = item_number_match.group(2) if item_number_match else None
    description = description_match.group(1) if description_match else None
    size = description_match.group(2) if description_match else None
    uom = uom_match.group(1) if uom_match else None
    quantity = quantity_match.group(1) if quantity_match else None
    unit_price = quantity_match.group(2) if quantity_match else None
    extended_price = extended_price_match.group(1) if extended_price_match else None

    # Print the extracted information
    print(f"{item_line}. {item_number} {description} {quantity} {unit_price} {extended_price}")

    # Return the extracted values as a dictionary
    return {
        "Item Line": item_line,
        "Item Number": item_number,
        "Description": description,
        "Quantity": quantity,
        "Unit Price": unit_price,
        "Extended Price": extended_price
    }

# Your modified output
output = extracted_output 

# Initialize an empty list to store the extracted data
item_data = []

# Process each item using the function and store the data in a list
for item in output:
    item_dict = process_item(item)
    item_data.append(item_dict)
```
